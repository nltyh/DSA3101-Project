```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
products = pd.read_csv("Product Table.csv")
```

Number of products per category

```python
sorted_order = products['Category'].value_counts().index

sns.set(rc = {'figure.figsize':(10,8)})
sns.countplot(y = 'Category', data = products, order = sorted_order)
plt.title('Number of products per Category')
plt.tick_params(axis = 'y', labelsize = 8)
```

Distribution of Product Prices

```python
sns.histplot(products['Product Price'], bins = 50)
plt.title('Distribution of Product Prices')
```

Exploring products with high price

```python
products[products['Product Price'] > 750]

# The products with high prices are laptops and elliptical machines, which are usually expensive
# Not outlier or errors 

# taking a better look at the distribution of product prizes < 750
df = products[products['Product Price'] < 750]

sns.histplot(df['Product Price'], bins = 10)
plt.title('Distribution of Product Prices')

# most products are less than $100, and the prices tapers down from $100 to $600
```

Mean prices of products in each category

```python
mean_prices = products.groupby('Category')['Product Price'].mean().reset_index()

mean_prices = mean_prices.sort_values(by='Product Price', ascending=False)

sns.barplot(x='Product Price', y='Category', data=mean_prices)

sns.set(rc = {'figure.figsize':(10,8)})
plt.tick_params(axis = 'y', labelsize = 8)
plt.title('Mean Product Prices by Category')
plt.xlabel('Mean Price')
plt.ylabel('Category')
```
