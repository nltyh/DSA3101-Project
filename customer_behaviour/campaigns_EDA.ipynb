```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import gdown
```

```python
file_id = '1GtTpo6aDmayZDTzaEbZSPLqvQ8riJN59' 
url = f'https://drive.google.com/uc?id={file_id}'
output = 'final_campaigns.csv'
gdown.download(url, output, quiet=False)

campaigns = pd.read_csv(output)
```

Overview of Data

```python
print("Dataset Overview:")
print(campaigns.info())

print("Missing Values:")
print(campaigns.isnull().sum())

print("Duplicate Rows:")
duplicate_count = campaigns.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

print("Descriptive Statistics:")
print(campaigns.describe(include='all'))
```

```python
print(campaigns.columns)
```

Distribution of Campaign Types 

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=campaigns, x='campaign_type')
plt.title('Distribution of Campaign Types')
plt.show()
```

Distribution of Channels

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=campaigns, x='channel')
plt.title('Distribution of Marketing Channels')
plt.show()
```

Distribution across Campaign Type and Channel 

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=campaigns, x='campaign_type', hue='channel')
plt.title('Count of Campaigns by Campaign Type and Channel')
plt.show()
```

Distribution of Subject Length 

```python
plt.figure(figsize=(8, 5))
sns.histplot(campaigns['subject_length'], kde=True, bins=30)
plt.title('Distribution of Subject Length in Campaigns')
plt.xlabel('Subject Length')
plt.show()
```

Analyse distribution of Subject Customisation by Channel

```python
feature_columns = [
    'subject_with_personalization', 'subject_with_emoji', 
    'subject_with_deadline', 'subject_with_bonuses', 
    'subject_with_discount', 'subject_with_saleout']


for feature in feature_columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=campaigns, x='channel', hue=feature)
    plt.title(f'Usage of {feature.replace("_", " ").title()} by Channel')
    plt.show()
```
